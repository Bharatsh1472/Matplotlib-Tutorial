# BAR PLOT
* In Matplotlib, a bar plot (or bar chart) is a chart that represents categorical data with rectangular bars. Each bar's height (or length if horizontal) represents the value of the data for that category. It's commonly used to compare quantities across categories.
* A basic bar plot in Matplotlib is created with the plt.bar() function, which requires:
  
  x-axis values (typically categorical data, such as labels),
  
  y-axis values (numerical data, often the counts or values associated with each category).

```python
import matplotlib.pyplot as plt

x = ["PART 1","PART 2","PART 3","PART 4","PART 5", "Part 6", "Part 7", "Part 8", "Part 9", "Part 10"]
y = [54,37,37,58,78,71,81,67,59,56]

# Defining the colors for the bars
colors = ["violet","indigo","blue","green","yellow","orange", "red"]

# Creating the bar plot, specifying the x-values, y-values, and the colors for the bars
plt.bar(x,y, color = colors, edgecolor="black") # edgecolor for the borders around the bars

# Labeling the x-axis and y-axis
plt.xlabel("Different Parts", fontsize = 13)
plt.ylabel("Popularity", fontsize = 15)

# Adding a title to the plot with a specified font size
plt.title("Popularity of Different Parts of Fast & Furious", fontsize = 16)

# Saving the plot as a PNG image file
plt.savefig("barplot.png")
plt.xticks(rotation=45)  #Rotate x-axis labels if they overlap
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/expense3.xlsx")
df = pd.DataFrame(data)
print(df)
print()
# Group the data by the "Payment Mode" column and calculate the SUM of the "Amount" for each payment mode
grp_by = df.groupby("Payment Mode")["Amount"].sum()

# Group the data by the "Payment Mode" column and calculate the mean of the "Amount" for each payment mode
# grp_by = df.groupby("Payment Mode")["Amount"].mean()

print(grp_by)

colors = ["violet","indigo","blue"]

# Create a bar plot with Payment Mode on the x-axis and the average Amount on the y-axis
plt.bar(grp_by.index, grp_by.values, color = colors, edgecolor="black")

# Labeling the x-axis and y-axis
plt.xlabel("Payment Mode", fontsize = 13)
plt.ylabel("Amount", fontsize = 15)

# Add text labels on top of each bar
for i, v in enumerate(grp_by.values):
    plt.text(i, v + max(grp_by.values)*0.01, str(v), ha='center', va='bottom')

plt.show()
```
