# LINE PLOT
* In Matplotlib, a line plot is a graph that displays data points connected by straight lines. Line plots are typically used to visualize data trends over a continuous interval, such as time, and are ideal for showing changes, patterns, or relationships in data.
* A line plot in Matplotlib is created using the plt.plot() function, where:

  x-axis values represent the independent variable (e.g., time, dates, categories).

  y-axis values represent the dependent variable (values that change in relation to the x-axis variable).

```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Define the x-axis data, representing the days of the week
x = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"]

# Define y-axis data for week 1, representing popularity scores
y = [300, 250, 290, 150, 225]

# Define y-axis data for week 2, representing popularity scores
y1 = [500, 220, 310, 190, 200]

# Plot data for week 1
plt.plot(x, y, marker="o", ls="--", color="red", label="week1")  # 'o' markers, dashed line style, red color, with label

# Plot data for week 2
plt.plot(x, y1, marker="o", ls="-", color="green", label="week2")  # 'o' markers, solid line style, green color, with label

# Display the legend to differentiate the two lines (week1 and week2)
plt.legend()

# Set the x-axis & y-axis labels with a specified font size
plt.xlabel("Weekdays", fontsize=12)
plt.ylabel("Popularity", fontsize=12)

plt.title("Analyzing Restaurant Popularity Across Different Days of the Week", fontsize=10)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/expense3.xlsx")
df = pd.DataFrame(data)
print(df)
print()

# Group the data by "Category" and calculate the sum of "Amount" for each category
grpby = df.groupby("Category")["Amount"].sum()
print(grpby)

# Plot the summed amount for each category as a line plot
plt.plot(grpby.index, grpby.values, marker="o", color="blue") 


plt.xlabel("Category", fontsize=12)
plt.ylabel("Amount", fontsize=12)
plt.title("Category-wise Spending Analysis", fontsize=14)

# Add data labels to each point on the plot
for i, value in enumerate(grpby.values):
    # plt.text(i, value, str(value), ha='center', va='bottom', fontsize=10)
    plt.text(i, value + max(grpby.values) * 0.01, str(value), ha='center', va='bottom', fontsize=10)

plt.show()
```
