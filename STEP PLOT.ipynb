# STEP PLOT

* A step plot is used to display data that changes at discrete intervals, making it suitable for visualizing stepwise or piecewise data.
* It connects data points with horizontal and vertical lines, emphasizing the transition points.
* In Matplotlib, you can create a step plot using the step() function.
* Step plots are ideal for data that doesn't change continuously, making them a powerful visualization tool for discrete datasets.

```python
import matplotlib.pyplot as plt

# Defining the x-axis labels (days of the week)
days = ["day1", "day2", "day3", "day4", "day5", "day6", "day7"]

# Defining the y-axis values (some measurement for each day)
x = [30, 40, 25, 30, 40, 25, 50]

# Creating a step plot with markers at each data point
plt.step(days, x, marker="o")  # The default `where` parameter is "pre", meaning steps are drawn before the data point.

# Uncomment one of the lines below to modify the step plot style:
# plt.step(days, x, where="post")  # Steps are drawn after the data point.
# plt.step(days, x, where="mid")   # Steps are drawn at the midpoint between consecutive data points.

# Displaying the plot
plt.show()
```

```python
import pandas as pd  
import matplotlib.pyplot as plt 

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/expense3.xlsx")
df = pd.DataFrame(data)

# Grouping the data by "Category" and calculating the total sum of "Amount" for each category
grp_by = df.groupby("Category")["Amount"].sum()
print(grp_by)

# Creating a step plot for the grouped data
plt.step(grp_by.index, grp_by.values, marker="*", where="mid")  # Adding star markers at midpoints

# Adding labels to the axes with font size 14
plt.xlabel("Category", fontsize=14)
plt.ylabel("Amount", fontsize=14)
plt.title("Total Expenses by Category", fontsize=16)

# Displaying the plot
plt.show()
```
