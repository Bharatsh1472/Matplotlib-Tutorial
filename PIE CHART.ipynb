# PIE CHART

* A pie chart is a circular statistical graphic which is divided into slices to illustrate numerical proportion. In Matplotlib, we use the plt.pie() function to create pie charts.
* The plt.pie() function requires:
  - Data: A sequence of numerical values representing the proportions of each slice.
  - Labels: A sequence of labels for each slice.

```python
import matplotlib.pyplot as plt

brands = ["Samsung","Apple","Xiaomi","Motorola","OnePlus","Sony"]
x = [22,36,45,14,25,13]
c = ["Violet","Indigo","Blue","Yellow","Orange","Red"]

# Explode parameter to detach a slice (Sony in this case)
ex = [0,0,0,0,0,0.1]
# plt.pie(x, labels = brands, colors = c, explode = ex, shadow = True, autopct='%.2f')
# plt.pie(x, labels = brands, colors = c, explode = ex, shadow = True, autopct='%.1f')
plt.pie(x, labels=brands, colors=c, explode=ex, shadow=True, autopct=lambda p: '{:.0f}'.format(p * sum(x) / 100))
# autopct: This is a function that determines how to format and display the percentage labels on each slice. 
# In this case, it calculates the percentage of the total market share for each slice and formats it as a string with one decimal place.
plt.title("Market Share of Smartphone Brands (2023)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/expense3.xlsx")
df = pd.DataFrame(data)
print(df)

# Group the data by "Payment Mode" and sum the "Amount" for each group
grpby = df.groupby("Payment Mode")["Amount"].sum()
print(grpby)

c = ["Yellow","Orange","Violet"]

plt.pie(grpby.values, labels = grpby.index, colors = c, autopct = "%.0f")
# Here, "%.0f" specifies that the percentage should be displayed as an integer (no decimal places).

plt.title("Expense Breakdown by Payment Method")
plt.show()
```
