# VIOLIN PLOT

* A violin plot in Matplotlib is a method of visualizing the distribution of a dataset. While a box plot shows summary statistics (like median and interquartile range), a violin plot also displays the kernel density estimation of the data, offering a clearer representation of its distribution.
* Violin plots are particularly useful for comparing distributions between multiple groups or datasets.
* The violinplot() function in Matplotlib can be used to create a violin plot. 

```python
import matplotlib.pyplot as plt
x = [12,20,35,30,25,19,35,40,70,75,41,55,70,85,80]
plt.violinplot(x, 
               showmedians=True,  # Show the median value as a horizontal line
               widths=0.8)  

# Add a title and axis labels
plt.title("Violin Plot of Data Distribution", fontsize=14)  # Title of the plot
plt.ylabel("Values", fontsize=12)  # Label for the y-axis
plt.xticks([1], ["Data"])  # Label the x-axis (single group in this case)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/ESD.xlsx")
df = pd.DataFrame(data)

# Group data by a categorical variable (e.g., Gender) and extract numerical values (e.g., Age)
groups = df.groupby("Gender")["Age"].apply(list)  # Replace "Gender" with the desired categorical column
print(groups)
group_labels = groups.index.tolist()  # Get the group labels
group_data = groups.tolist()  # Get the numerical data for each group

# Create the violin plot
plt.figure(figsize=(10, 6))  
plt.violinplot(group_data, 
               showmedians=True,  # Display medians in the plot
               widths=0.8) 

# Add labels and titles
plt.title("Distribution of Age by Gender", fontsize=14)  
plt.ylabel("Age", fontsize=12)  
plt.xlabel("Gender", fontsize=12)  
plt.xticks(range(1, len(group_labels) + 1), group_labels, fontsize=10)  # Add group labels on the x-axis

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
