# STEM PLOT

* A stem plot is used to visualize data as vertical lines extending from a baseline, ending in markers to represent individual data points. It's particularly useful for displaying discrete data or emphasizing individual data points in a sequence.
* In Matplotlib, you can create a stem plot using the stem() function.
* Key Parameters of stem():
  - linefmt: Sets the style of the vertical lines (e.g., '--' for dashed lines).
  - markerfmt: Defines the style of the markers (e.g., 'o' for circles).
  - basefmt: Configures the style of the baseline (e.g., 'r-' for a red solid line).


```python
import matplotlib.pyplot as plt

x = [10,20,50,30,40,50,60,40,70,60]

plt.stem(x, linefmt = "--", markerfmt = "D")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/ESD.xlsx")
df = pd.DataFrame(data)
df2 = df.head(50)

plt.figure(figsize=(10, 6)) 

# plt.stem(df2["Age"], bottom = 20)
# plt.plot(df2["Age"])

# Create a stem plot for "Age"
plt.stem(df2.index,df2["Age"], linefmt="C0-", markerfmt="C0o", basefmt="r-", bottom=20, label="Age (Stem Plot)")

# Overlay a line plot for "Age"
plt.plot(df2.index, df2["Age"], linestyle="--", color="purple", label="Age (Line Plot)")

# Add a grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add title and labels
plt.title("Stem and Line Plot of Age for First 50 Employees", fontsize=14)
plt.xlabel("Employee Index", fontsize=12)
plt.ylabel("Age", fontsize=12)

# Add a legend to distinguish between the plots
plt.legend()

plt.show()
```
