# HISTOGRAM 

* A histogram is a type of bar plot that shows the distribution of a dataset. It divides data into bins (intervals) and displays the frequency of data points in each bin.

```python
import matplotlib.pyplot as plt

x = [12,2,15,26,34,28,49,9,32,9,2,48,6,5,62,56,26,2]

# Create the histogram
plt.figure(figsize=(8, 6))  # Set figure size for better visibility
plt.hist(x, 
         bins=10,  # Divide the data into 10 equal-width intervals
         edgecolor="black",  # Add black edges to each bar for clarity
         color="purple",  # Set the fill color of the bars to purple
         alpha=0.8)  # Adjust transparency for better visibility

# Add title and axis labels
plt.title("Histogram of Data Distribution", fontsize=14) 
plt.xlabel("Data Range (Values)", fontsize=12) 
plt.ylabel("Frequency (Counts)", fontsize=12) 

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/ESD.xlsx")
df = pd.DataFrame(data)
print(df)

# plt.hist(df["Age"], bins = 25)

plt.figure(figsize=(8, 6))
plt.hist(df["Annual Salary"], 
         bins=50,  # Number of bins (adjusted to reduce congestion)
         density=False,  # Frequency count, not density
         histtype='step',  # Step outline histogram
         color="violet",  # Custom color
         linewidth=2)  # Thickness of the outline

# Add titles and labels
plt.title("Distribution of Annual Salaries", fontsize=14) 
plt.xlabel("Annual Salary", fontsize=12)  
plt.ylabel("Frequency", fontsize=12) 

# Customize gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
