# SCATTER PLOT

* In Matplotlib, a scatter plot is a type of plot that displays individual data points on a two-dimensional graph. Scatter plots are useful for showing the relationship between two variables and identifying patterns, trends, or correlations within data. Each point represents an observation with values on the x and y axes.
* A scatter plot in Matplotlib is created using the plt.scatter() function. This function requires:
  - x-axis values: Independent variable values.
  - y-axis values: Dependent variable values.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 50 random integer values from 1 to 10 for the x-axis
x = np.random.randint(1,10,50)

# Generate 50 random integer values from 10 to 100 for the y-axis
y = np.random.randint(10,100,50)

# Generate random values from 10 to 100 to represent color intensities for each point
color = np.random.randint(10,100,50)

# Generate random values from 10 to 100 to represent sizes of each point
size = np.random.randint(10,100,50)

# plt.scatter(x,y,marker = "^", cmap = "any", c = color, s = size) 
# Note: 'any' will give u the range of colors from which u can pick any color of your choice
# plt.scatter(x,y,marker = "^", cmap = 'cividis', c = color, s = size)
# plt.scatter(x,y,marker = "^", cmap = "viridis", c = color, s = size)
# plt.scatter(x,y,marker = "^", cmap = "coolwarm", c = color, s = size)
plt.scatter(x,y,marker = "^", cmap = "inferno", c = color, s = size)

# Add a color bar to show the scale of color intensity
plt.colorbar(label="Color Intensity")

# Set x-axis and y-axis labels
plt.xlabel("X-axis Values (1-10)")
plt.ylabel("Y-axis Values (10-100)")

plt.title("Scatter Plot of Random Values with Variable Size and Color Intensity")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/ESD.xlsx")
df = pd.DataFrame(data)
print(df)

# Scatter plot for the entire dataset

# Increase the figure size for better readability
plt.figure(figsize=(10, 6))

# Create a scatter plot with "Annual Salary" on the x-axis and "EEID" on the y-axis
# The size of each point is based on "Age", and we use a colormap to indicate "Age"
plt.scatter(df["Annual Salary"], df["EEID"], s=df["Age"] * 5, c=df["Age"], cmap="viridis", alpha=0.5)

plt.colorbar(label="Age")   # Add a color bar to indicate the age values (based on the color)

plt.xlabel("Annual Salary")
plt.ylabel("Employee ID (EEID)")

plt.title("Scatter Plot of Annual Salary vs. Employee ID (All Data)")
plt.show()

# Scatter plot for the first 30 rows

# Select the first 30 rows
df_first_30 = df.head(30)

plt.figure(figsize=(10, 6))
plt.scatter(df_first_30["Annual Salary"], df_first_30["EEID"], s=df_first_30["Age"] * 10, c=df_first_30["Age"], cmap="viridis", alpha=0.7)
plt.colorbar(label="Age")
plt.xlabel("Annual Salary")
plt.ylabel("Employee ID (EEID)")
plt.title("Scatter Plot of Annual Salary vs. Employee ID (First 30 Rows)")
plt.show()
```
