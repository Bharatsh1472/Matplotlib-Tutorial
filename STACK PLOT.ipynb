# STACK PLOT

* A stack plot is used to visualize data that accumulates over time or categories.
* It shows the contribution of individual components to a whole, represented by "stacking" their values on top of each other.
* Stack plots are commonly used for displaying data like population trends, financial data, or energy production.
* In Matplotlib, you can create a stack plot using the stackplot() function.

```python
import matplotlib.pyplot as plt

# Define the x-axis values representing days
days = [1, 2, 3, 4, 5, 6, 7]

# Define the data for the number of participants (NOP) for three weeks
NOP1 = [5, 10, 30, 20, 35, 60, 80]   # Week 1 data
NOP2 = [10, 20, 30, 25, 40, 50, 90]  # Week 2 data
NOP3 = [8, 30, 50, 60, 70, 90, 100]  # Week 3 data

# Create a stack plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.stackplot(days, NOP1, NOP2, NOP3, labels=["Week 1", "Week 2", "Week 3"], colors=["skyblue", "lightgreen", "salmon"])

# Add a legend to the plot
plt.legend(loc="upper left", shadow=True, labelcolor="purple")

# Add labels and title
plt.title("Participation Trends Over a Week", fontsize=14)  # Add a title
plt.xlabel("Days", fontsize=12)  # Label for the x-axis
plt.ylabel("Number of Participants", fontsize=12)  # Label for the y-axis

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/ESD.xlsx")
df = pd.DataFrame(data)

# Display the first 50 rows for clarity
df2 = df.head(50)

# Aggregate data: Group by 'Country' and calculate total 'Annual Salary'
country_salary = df2.groupby("Country")["Annual Salary"].sum()

# Extract x-axis (countries) and y-axis (salaries)
countries = country_salary.index
salaries = country_salary.values

# Create a stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(range(len(countries)), salaries, labels=["Annual Salary"], colors=["green"])

# Add legend, title, and labels
plt.legend(loc="upper left", shadow=True, labelcolor="purple")
plt.title("Stack Plot of Annual Salary by Country", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Total Annual Salary", fontsize=12)

# Replace x-axis ticks with country names
plt.xticks(ticks=range(len(countries)), labels=countries)

# Show the plot
plt.show()
```
